```python
import pandas as pd
from collections import defaultdict
from collections import Counter
import re
import importlib
importlib.reload(pp)
import src.services.preprocessing as pp
```

```python
cnt = Counter()

# Identify the placeholders to be replaced
df = pd.read_csv(
    "../data/raw/Bitext_Sample_Customer_Support_Training_Dataset_27K_responses-v11.csv"
)

pattern = r"\{\{([^}]+)\}\}"
for text in pd.concat([df["instruction"], df["response"]]):
    matches = re.findall(pattern, text)
    for key in matches:
        cnt[key.strip()] += 1

print("Identified placeholders and their counts:")
for key, count in cnt.items():
    print(f"  {key}: {count}")
```
